## **Multi-Class Classification Example (.csv file)**

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Step 2: Load the dataset
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('iris.csv')
```

```python
# Step 3: Prepare features and labels
X = df.drop('variety', axis=1).values
y = df['variety'].values
y
```

```python
# Encode labels (convert species to integers)
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
y_encoded
```

```python
# Convert labels to one-hot encoding
y_onehot = tf.keras.utils.to_categorical(y_encoded)
y_onehot
```

```python
# Step 4: Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_onehot, test_size=0.2, random_state=42)
```

```python
# Step 5: Build the neural network model
model = Sequential([
    Dense(10, input_shape=(4,), activation='relu'),
    Dense(8, activation='relu'),
    Dense(3, activation='softmax')  # 3 output classes
])
model.summary()
```

```python
# Step 6: Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Step 7: Train the model
model.fit(X_train, y_train, epochs=5, batch_size=5, verbose=1, shuffle = True) # By default, model.fit(...) uses shuffle=True (unless you set shuffle=False)
```

```python
# Step 8: Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {accuracy:.2f}")
```
