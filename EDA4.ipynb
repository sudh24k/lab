# Task
Load and standardize a sample dataset, then apply PCA. Generate a scree plot and a cumulative variance plot to identify the minimum number of principal components needed to retain at least 90% of the total variance, and summarize these findings.

## Load and Standardize Data

### Subtask:
Load a sample dataset and standardize it using StandardScaler, which is a necessary preprocessing step for PCA.


**Reasoning**:
First, I'll import pandas for DataFrame manipulation and StandardScaler for data standardization. Then, I'll load the Iris dataset from scikit-learn, convert it into a pandas DataFrame named `df`, and proceed to standardize it using `StandardScaler`. The standardized data will be stored in `scaled_data`.



```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

print("Original DataFrame head:")
print(df.head())

scaler = StandardScaler()
scaler.fit(df)

scaled_data = scaler.transform(df)

df_scaled = pd.DataFrame(scaled_data, columns=df.columns)

print("\nStandardized DataFrame head:")
print(df_scaled.head())

print("Data loaded and standardized successfully.")
```

## Apply PCA

### Subtask:

## Apply PCA

### Subtask:
Apply Principal Component Analysis to the standardized dataset to transform it into its principal components.

#### Instructions
1. Import the `PCA` class from `sklearn.decomposition`.
2. Instantiate a `PCA` object. Do not specify `n_components` at this stage, so that all principal components are computed.
3. Fit the PCA model to the standardized DataFrame `df_scaled`.
4. Transform the `df_scaled` data using the fitted PCA model to get the principal components. Store the result in a variable, for example, `principal_components`.
5. Store the explained variance ratio of each principal component, which can be accessed via the `.explained_variance_ratio_` attribute of the fitted PCA model.

**Reasoning**:
Now, I will implement the PCA steps outlined in the previous markdown, which include importing PCA, instantiating it, fitting and transforming the standardized data, and then extracting the explained variance ratios.



```python
from sklearn.decomposition import PCA

pca = PCA()

pca.fit(df_scaled)

principal_components = pca.transform(df_scaled)

explained_variance_ratio = pca.explained_variance_ratio_

print("PCA applied successfully.")
print("Shape of principal components array:", principal_components.shape)
print("Explained variance ratio per component:", explained_variance_ratio)
```

## Generate Scree Plot and Cumulative Variance Plot

### Subtask:
Generate two plots: a scree plot showing the explained variance ratio for each principal component, and a cumulative explained variance plot to visualize the total variance retained as more components are added. Include appropriate labels and legends for clarity.


**Reasoning**:
I will generate the scree plot and the cumulative variance plot as described in the instructions. This involves importing `matplotlib.pyplot` and `numpy`, then plotting the explained variance ratio for the scree plot and the cumulative explained variance for the second plot, adding appropriate labels, titles, and a reference line for 90% variance.



```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Create a scree plot
plt.figure(figsize=(10, 5))
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, color='skyblue')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.xticks(range(1, len(explained_variance_ratio) + 1))
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Calculate the cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# 3. Create a cumulative variance plot
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='-', color='purple')
plt.axhline(y=0.90, color='r', linestyle='--', label='90% Variance Retained')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance')
plt.xticks(range(1, len(cumulative_explained_variance) + 1))
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

print("Scree plot and Cumulative Variance plot generated successfully.")
```
