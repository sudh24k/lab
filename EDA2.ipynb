## Formulate Hypotheses on Missingness Mechanism

### Subtask:
For at least two variables with a substantial number of missing values, formulate a hypothesis about the mechanism of missingness (MCAR, MAR, or MNAR). Provide statistical evidence to support your hypotheses, potentially using statistical tests or comparisons between groups with and without missing data.


**Reasoning**:
To begin analyzing the missingness mechanism for the 'age' variable, I will first split the DataFrame into two groups: one where 'age' values are missing and another where they are present. This separation is crucial for conducting comparative statistical tests as outlined in the instructions.



```python
import numpy as np
from scipy.stats import chi2_contingency, ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

missing_cols = ['age', 'deck', 'embarked', 'embark_town']

age_missing = df[df['age'].isnull()]
age_present = df[df['age'].notnull()]

print(f"Number of rows where 'age' is missing: {len(age_missing)}")
print(f"Number of rows where 'age' is present: {len(age_present)}")
```

**Reasoning**:
To compare the distribution of 'pclass' between the 'age' missing and 'age' present groups, I will create a contingency table for 'pclass' from both subsets and perform a chi-squared test for independence, as specified in the instructions.



```python
print("\n--- Comparing 'pclass' distribution for 'age' missing vs. 'age' present ---")
contingency_table_pclass = pd.crosstab(df['pclass'], df['age'].isnull())

chi2, p, dof, ex = chi2_contingency(contingency_table_pclass)

print("Contingency Table (pclass vs. Age Missing Status):\n", contingency_table_pclass)
print(f"Chi-squared statistic: {chi2:.2f}")
print(f"P-value: {p:.3f}")

if p < 0.05:
    print("Conclusion: There is a statistically significant difference in 'pclass' distribution between passengers with missing age and those with present age. (Suggests MAR or MNAR)")
else:
    print("Conclusion: No statistically significant difference in 'pclass' distribution between passengers with missing age and those with present age. (Consistent with MCAR)")

plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='pclass', hue=df['age'].isnull(), palette='coolwarm')
plt.title('Pclass Distribution by Age Missing Status')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.legend(title='Age Missing', labels=['Present', 'Missing'])
plt.show()
```

**Reasoning**:
To continue analyzing the missingness mechanism for the 'age' variable, I will now compare the distribution of 'sex' between the 'age' missing and 'age' present groups using a chi-squared test and visualize the results, as specified in the instructions.



```python
print("\n--- Comparing 'sex' distribution for 'age' missing vs. 'age' present ---")
contingency_table_sex = pd.crosstab(df['sex'], df['age'].isnull())

chi2_sex, p_sex, dof_sex, ex_sex = chi2_contingency(contingency_table_sex)

print("Contingency Table (Sex vs. Age Missing Status):\n", contingency_table_sex)
print(f"Chi-squared statistic: {chi2_sex:.2f}")
print(f"P-value: {p_sex:.3f}")

if p_sex < 0.05:
    print("Conclusion: There is a statistically significant difference in 'sex' distribution between passengers with missing age and those with present age. (Suggests MAR or MNAR)")
else:
    print("Conclusion: No statistically significant difference in 'sex' distribution between passengers with missing age and those with present age. (Consistent with MCAR)")

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='sex', hue=df['age'].isnull(), palette='coolwarm')
plt.title('Sex Distribution by Age Missing Status')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Age Missing', labels=['Present', 'Missing'])
plt.show()
```

**Reasoning**:
To further analyze the missingness mechanism for the 'age' variable, I will compare the 'fare' distribution between the 'age' missing and 'age' present groups using an independent t-test and visualize the distributions with box plots, as instructed.



```python
print("\n--- Comparing 'fare' distribution for 'age' missing vs. 'age' present ---")

stat, p_fare = ttest_ind(age_present['fare'], age_missing['fare'])

print(f"T-statistic for fare: {stat:.2f}")
print(f"P-value for fare: {p_fare:.3f}")

if p_fare < 0.05:
    print("Conclusion: There is a statistically significant difference in 'fare' distribution between passengers with missing age and those with present age. (Suggests MAR or MNAR)")
else:
    print("Conclusion: No statistically significant difference in 'fare' distribution between passengers with missing age and those with present age. (Consistent with MCAR)")

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x=df['age'].isnull(), y='fare', palette='viridis')
plt.title('Fare Distribution by Age Missing Status')
plt.xlabel('Age Missing Status')
plt.ylabel('Fare')
plt.xticks(ticks=[0, 1], labels=['Present', 'Missing'])
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(age_present['fare'], color='blue', label='Age Present', kde=True, stat='density', alpha=0.5)
sns.histplot(age_missing['fare'], color='red', label='Age Missing', kde=True, stat='density', alpha=0.5)
plt.title('Fare Distribution for Age Present vs. Age Missing')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### Hypothesis for Missingness of 'age'

**Mechanism:** Missing At Random (MAR)

**Reasoning:**

The statistical tests revealed the following:

*   **'pclass' vs. 'age' missing status:** The chi-squared test yielded a p-value of 0.000, which is significantly less than 0.05. This indicates a statistically significant difference in the distribution of `pclass` between passengers with missing `age` and those with present `age`. Specifically, a higher proportion of passengers in `pclass` 3 have missing `age` compared to `pclass` 1 and 2.
*   **'sex' vs. 'age' missing status:** The chi-squared test yielded a p-value of 0.119, which is greater than 0.05. This indicates no statistically significant difference in the distribution of `sex` between passengers with missing `age` and those with present `age`.
*   **'fare' vs. 'age' missing status:** The independent t-test yielded a p-value of 0.003, which is significantly less than 0.05. This indicates a statistically significant difference in the mean `fare` paid by passengers with missing `age` compared to those with present `age`. The `fare` distribution for passengers with missing `age` appears to be lower than for those with present `age`.

Since the missingness of `age` is related to other *observed* variables (`pclass` and `fare`) in the dataset, but not to the `age` itself (if it were observed), the most appropriate missingness mechanism is **Missing At Random (MAR)**. If the missingness were related to `age` values themselves (e.g., older people tend to not report their age), it would be MNAR.
